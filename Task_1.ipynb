```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', skiprows=4)
```

```python
df.describe()
```

```python
df.info()
```

```python
df_2022 = df[['Country Name', '2022']].dropna()
df_2022 = df_2022.sort_values(by='2022', ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(df_2022['Country Name'], df_2022['2022'], color='lightblue')
plt.xlabel("Population (2022)")
plt.title("Top 10 Most Populous Countries (2022)")
plt.gca().invert_yaxis()  # highest at the top
plt.tight_layout()
plt.show()
```

```python
# Bottom 10 countries by population
df_bottom10 = df[['Country Name', '2022']].dropna().sort_values(by='2022', ascending=True).head(10)

plt.figure(figsize=(10, 6))
plt.barh(df_bottom10['Country Name'], df_bottom10['2022'], color='pink')
plt.xlabel("Population (2022)")
plt.title("10 Least Populous Countries (2022)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
regions = {
    'Asia': ['India', 'China', 'Indonesia', 'Pakistan', 'Bangladesh', 'Japan'],
    'Africa': ['Nigeria', 'Ethiopia', 'Egypt', 'South Africa'],
    'Europe': ['Germany', 'France', 'United Kingdom', 'Italy', 'Spain'],
    'Americas': ['United States', 'Brazil', 'Mexico', 'Canada', 'Argentina']
}

# Calculate total population per region
region_data = {}
for region, countries in regions.items():
    total = df[df['Country Name'].isin(countries)]['2022'].dropna().sum()
    region_data[region] = total

# Plot
plt.figure(figsize=(8, 5))
plt.bar(region_data.keys(), region_data.values(), color='darkcyan')
plt.ylabel("Population (2022)")
plt.title("Total Population by Region (Sample Countries)")
plt.tight_layout()
plt.show()
```

```python
# Filter only required columns
df_growth = df[['Country Name', '2000', '2022']].dropna()
df_growth['Growth'] = df_growth['2022'] - df_growth['2000']

# Top 10 countries with highest growth
df_top_growth = df_growth.sort_values(by='Growth', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(df_top_growth['Country Name'], df_top_growth['Growth'], color='mediumvioletred')
plt.xlabel("Population Growth (2000 to 2022)")
plt.title("Top 10 Countries with Highest Population Growth")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
df_top15 = df.sort_values(by='2022', ascending=False).head(15)

plt.figure(figsize=(10, 6))
plt.bar(df_top15['Country Name'], df_top15['2022'], color='darkorange')
plt.yscale('log')
plt.xticks(rotation=90)
plt.ylabel("Population (log scale)")
plt.title("Top 15 Countries by Population (Log Scale)")
plt.tight_layout()
plt.show()
```
